```python
import pandas as pd
import tiktoken
import os
os.listdir()
```

```python
import tiktoken
def count_tokens(text, model="gpt-3.5"):

    encoding = tiktoken.encoding_for_model(model)
    return len(encoding.encode(text))
# 示例文本
text="你好，最近怎么样？"
token_count = count_tokens(text)
print(f"Token 数量:{token_count}")
```

```python
for i in range(4):
    df=pd.read_csv(os.listdir()[i])
    df = df[df['label']==1]
    df['source_tokens']=df['source_text'].apply(count_tokens)
    df['target_tokens']=df['target_text'].apply(count_tokens)
    print(os.listdir()[i].split('_')[1],'nl_tokens:',round(df['source_tokens'].values.mean(),2))
    print(os.listdir()[i].split('_')[1],'pl_tokens:',round(df['target_tokens'].values.mean(),2))

```
